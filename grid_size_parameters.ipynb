```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)
```

```python
grid5 = KKAN_Convolutional_Network(grid_size=5)
grid10 = KKAN_Convolutional_Network(grid_size=10)
grid15 = KKAN_Convolutional_Network(grid_size=15)

grid20= KKAN_Convolutional_Network(grid_size=20)
grids = [grid5,grid10,grid15,grid20]
grid_sizes = [5,10,15,20]

params = [count_parameters(i) for i in grids]
```
